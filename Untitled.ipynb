```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exemple : Charger un DataFrame Ã  partir d'un fichier CSV
df = pd.read_csv('train_u6lujuX_CVtuZ9i.csv')

# Visualisation de l'histogramme
sns.histplot(data=df, x='LoanAmount', bins=20, kde=True, color='blue')
plt.xlabel('LoanAmount')
plt.ylabel('Credit_History')
plt.title('Histogramme de LoanAmount')
plt.show()
```

```python

# Box Plot pour 'ApplicantIncome' et 'CoapplicantIncome'
sns.boxplot(x='variable', y='value', data=pd.melt(df[['ApplicantIncome', 'CoapplicantIncome']]))
plt.xlabel('Income Type')
plt.ylabel('Income Value')
plt.title('Box Plot for ApplicantIncome and CoapplicantIncome')
plt.show()
```

```python
# Nuage de points entre 'LoanAmount' et 'ApplicantIncome'
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=df, color='green')
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
plt.title('Scatter Plot between ApplicantIncome and LoanAmount')
plt.show()
```

```python
# Heatmap pour la corrÃ©lation entre les variables numÃ©riques
correlation_matrix = df[['LoanAmount', 'ApplicantIncome', 'CoapplicantIncome']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Supposons que 'df' soit dÃ©jÃ  dÃ©fini

# SÃ©lectionnez la colonne 'Credit_History' pour le Pie Chart
credit_history_counts = df['Credit_History'].value_counts()

# CrÃ©ez un Pie Chart
labels = credit_history_counts.index
sizes = credit_history_counts.values
colors = ['lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Assure que le Pie Chart est un cercle.
plt.title('RÃ©partition des catÃ©gories de crÃ©dit')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('train_u6lujuX_CVtuZ9i.csv')
sns.boxplot(x='variable', y='value', data=pd.melt(df[['LoanAmount', 'ApplicantIncome', 'CoapplicantIncome']]))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Supposons que 'df' soit dÃ©jÃ  dÃ©fini

# SÃ©lectionnez la colonne 'Credit_History' pour le Pie Chart
credit_history_counts = df['Loan_Status'].value_counts()

# CrÃ©ez un Pie Chart
labels = credit_history_counts.index
sizes = credit_history_counts.values
colors = ['green', 'blue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Assure que le Pie Chart est un cercle.
plt.title('RÃ©partition des catÃ©gories de crÃ©dit')
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
p = figure(x_range=df['Loan_Status'],title='Diagramme en barres interactif')
p.vbar(x=df['Loan_Status'], top=df['ApplicantIncome'], width=0.9)
show(p)


#x='Loan_Status', y='ApplicantIncome'
```

```python
from bokeh.models import HoverTool

p = figure( title='Nuage de points interactif')
p.scatter(x='Loan_Status', y='ApplicantIncome', size='Value', color='Category', source=df)

hover = HoverTool()
hover.tooltips = [('Category', '@Category'), ('Value', '@Value')]
p.add_tools(hover)

show(p)
```
